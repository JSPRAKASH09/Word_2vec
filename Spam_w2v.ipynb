```python
import gensim
import numpy as np
from gensim.parsing.preprocessing import remove_stopwords
import pandas as pd
import nltk
from sklearn.model_selection import train_test_split
```

```python
dt=pd.read_csv("spam.csv",encoding='latin-1')
pd.set_option('display.max_colwidth',100)
dt=dt.drop(labels=["Unnamed: 3","Unnamed: 4", "Unnamed: 2"],axis=1)
dt.columns=["lables","Text"]
```

```python
dt
```

```python
dt["text_clean"]=dt["Text"].apply(lambda x: gensim.utils.simple_preprocess(x))
dt.head()
```

```python
import gensim.downloader as api
wiki_embeddings=api.load('glove-wiki-gigaword-100')
```

```python
x_train,x_test,y_train,y_test=train_test_split(dt['text_clean'],dt['lables'],test_size=0.2)
```

```python
w2v_model=gensim.models.Word2Vec(x_train,vector_size=100,window=5,min_count=2)
```

```python
w2v_model.wv['king']
```

```python
w2v_model.wv.most_similar('king')
```

```python
w2v_model.wv.index_to_key
```

```python
word_vect=np.array([np.array([w2v_model.wv[i] for i in ls if i in w2v_model.wv.index_to_key])
                    for ls in x_test])
```

```python
for i,v in enumerate(word_vect):
    print(len(x_test.iloc[i]),len(v))
```

```python
word_vect_avg=[]

for vect in word_vect:
    if len(vect)!=0:
        word_vect_avg.append(vect.mean(axis=0))
    else:
        word_vect_avg.append(np.zeros(100))
```

```python
for i,v in enumerate(word_vect_avg):
    print(len(x_test.iloc[i]),len(v))
```
